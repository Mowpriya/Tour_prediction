## Handling Missing data

* Replaced missing data of columns:
*  Age, DurationOfPitch,NumberOFTrips,NumberOfChildrenVisiting and MonthlyIncome with median of respective columns
*  Dropped missing values of remaining columns rowwise using dropna

```python
a=['Age','DurationOfPitch','NumberOfTrips','NumberOfChildrenVisiting','MonthlyIncome']
for i in a:
    df[i].fillna(df[i].median(),inplace=True)
```

```python
df.dropna(axis=0,inplace=True)
```
